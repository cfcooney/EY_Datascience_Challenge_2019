```python
import pandas as pd
import numpy as np
import math

#####Load in main training dataset#####
df = pd.read_csv('data_train/data_train.csv')
df.set_index("hash", inplace=True)
df.drop(["Unnamed: 0"], axis=1, inplace=True)
#df.set_index("trajectory_id", inplace=True)  
df.head(15)
```

```python
def calculateDistance_fromCC(x2,y2):
    
    """
    Compute Distance between starting points and CC.
    """
    CC_MID_X = 3760901.5068
    CC_MID_Y = -19238905.6133
    dist = math.sqrt((x2 - CC_MID_X)**2 + (y2 - CC_MID_Y)**2)  
    return dist

df["dist"] = list(map(calculateDistance_fromCC,df["x_entry"],df["y_entry"]))
```

```python
"""
Compute the net trajectory and the most recent trajectory
for each of the GPS Ids. Assigned here to each tag but 
only used in the final location data.
"""
def net_trajectory(distances):
    
    """
    Computes the net trajectory and most recent trajectory in
    relation to centre-point.
    """
    
    traj =[]
    if len(distances) >= 2:
        for i in range(len(distances)-1):
            traj.append(distances[i] - distances[i+1])

        recent_traj = distances[-2] - distances[-1]
        return np.sum(traj), recent_traj
    else:
        return 0.0, 0.0
    

"""
Compute mean journey distance.
"""
def calculateDistance(x1,y1,x2,y2):
    
    """
    Compute Distance between two points.
    """
    dist = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)  
    return dist


unique_ids = df.index.unique()
print(unique_ids.shape)

import time
start = time.time()
for i in unique_ids[:100]:

    #####trajectories#####
    distances = df[df.index == i]["dist"].values
    net_tr, prev_tr = net_trajectory(distances)
    df.loc[[i],'net_tr'] = net_tr
    df.loc[[i],'prev_tr'] = prev_tr
    
    #####mean distances#####
    x_entry = df[df.index == i]["x_entry"].values
    y_entry = df[df.index == i]["y_entry"].values
    x_exit = df[df.index == i]["x_exit"].values
    y_exit = df[df.index == i]["y_exit"].values
    
    mean_dist = []
    for w,x,y,z in zip(x_entry,y_entry,x_exit,y_exit):
        mean_dist.append(calculateDistance(w,x,y,z))
    
    df.loc[[i],'mean_dist'] = np.mean(mean_dist)
print(f"run time: {time.time()-start} seconds")
```

```python
"""
Load pickle containing dicts computed in cluster using functions above
and convert into df containing new values.
"""

import pickle
with open('filename.pickle', 'rb') as handle:
    data = pickle.load(handle)
    
df = pd.DataFrame(data=None, columns=pd.DataFrame.from_dict(data[0]).columns.values)
for d in data:
    new_df = pd.DataFrame.from_dict(d)
    new_df = new_df[new_df["mean_dist"].notnull()]
    df = pd.concat([df,new_df])
```

```python
"""
Loading preprocessed features from csv files and adding to
main df.
"""
df_home = pd.read_csv('data_train/home_features.csv')
df_nj = pd.read_csv('data_train/nj.csv')

df["x_home"] = df_home["x_home"]
df["y_home"] = df_home["y_home"]
df["nj"] = df_nj["n_j"]
```

```python
final_df = pd.DataFrame(data=None, columns=df.columns.values)

for n,i in enumerate(unique_ids):
    final_df.loc[n] = df[df.index == i].iloc[-1]
final_df.set_index("trajectory_id", inplace=True)   
```

```python
"""
Computations which only require the final location data.
"""

def journey_time(x,y):
    """
    Compute journey time in seconds.
    """
    x = pd.to_datetime(x)
    y = pd.to_datetime(y)
    return (y-x).total_seconds()

final_df["j_time"] = list(map(journey_time, final_df["time_entry"], final_df["time_exit"])) 
```
