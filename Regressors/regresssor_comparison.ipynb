```python
"""
GPS features computed previously. Several regression techniques compared.
"""
import pandas as pd
import numpy as np
import math
from sklearn.metrics import accuracy_score, f1_score
df = pd.read_csv('data_train/4_features.csv')
df_home = pd.read_csv('data_train/home_features.csv')
df_speed = pd.read_csv('data_train/data_train.csv')
df.set_index("trajectory_id", inplace=True)
df.tail(20)
```

```python
def calculateDistance(x2,y2):
    
    """
    Compute Distance between starting points and CC.
    """
    CC_MID_X = 3760901.5068
    CC_MID_Y = -19238905.6133
    dist = math.sqrt((x2 - CC_MID_X)**2 + (y2 - CC_MID_Y)**2)  
    return dist

def accuracy_f_score(y_pred,y_true):
    print(f"Accuracy score: {round(accuracy_score(y_true, y_pred) * 100,2)}%")
    print('\033[92m' + f"F1 score: {round(f1_score(y_true, y_pred) * 100,2)}%" + '\033[0m')
    return round(accuracy_score(y_true, y_pred) * 100,2), round(f1_score(y_true, y_pred) * 100,2)
    

def in_city(y_pred):
    targets = []
    for pred in y_pred:
        if (3750901.5068 <= pred[0] <= 3770901.5069) and (-19268905.6133 <= pred[1] <= -19208905.6133):
            targets.append(1)
        else:
            targets.append(0)
    return targets
```

```python
df["x_home"] = df_home["x_home"].values
df["y_home"] = df_home["y_home"].values
df["vel"] = df_speed["vmean"].values
df["vel"] = df["vel"].fillna((df["vel"].mean()))
df["dist"] = list(map(calculateDistance,df["x_entry"],df["y_entry"]))
```

```python
"""
Store/Load dataset and split
"""

df.to_csv('data_train/multi_features.csv', index=True)
df = pd.read_csv('data_train/multi_features.csv') 
df.set_index("trajectory_id", inplace=True)
df_train = df
y = [df["x_exit"].values, df["y_exit"].values]
y = np.transpose(np.array(y))
df_train.drop(["Tag_id","n_j","vel","x_exit","y_exit"], axis=1, inplace=True)
X = df_train.values
df_train.head()
```

```python
"""
Import some Regression algorithms to test on the dataset.
"""
import sklearn
from sklearn.multioutput import MultiOutputRegressor as MOR
from sklearn.linear_model import SGDRegressor, LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge, Perceptron
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor as RdF
import xgboost as xgb
```

```python
""" XGBoost Regression"""
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=410,shuffle=True)
data_dmatrix = xgb.DMatrix(data=x_train,label=y_train)
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree=1.0,colsample_bylevel=1, learning_rate=1.0,
                max_depth=15, alpha=0.1, n_estimators=25, reg_alpha=0, reg_lambda=0.0,gamma=0.001)

clf = MOR(xg_reg)

clf.fit(x_train,y_train)
pred = clf.predict(x_test)

pred = in_city(pred) # convert to binary predictions
y_test = in_city(y_test)

xg_acc, xg_f1 = accuracy_f_score(pred,y_test)
```

```python
""" Random Forest Regression"""
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=410,shuffle=True)
clf = MOR(RdF(n_estimators=50, min_samples_split=2, min_samples_leaf=2))
clf.fit(x_train,y_train)
pred = clf.predict(x_test)

pred = in_city(pred) # convert to binary predictions
y_test = in_city(y_test)

rdf_acc, rdf_f1 = accuracy_f_score(pred,y_test)
```

```python
""" Bayesian Ridge Regression"""
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=410,shuffle=True)
clf = MOR(BayesianRidge(n_iter=500, alpha_1=0.01, alpha_2=0.01))
clf.fit(x_train,y_train)
pred = clf.predict(x_test)

pred = in_city(pred) # convert to binary predictions
y_test = in_city(y_test)

brr_acc, brr_f1 = accuracy_f_score(pred,y_test)
```

```python
""" Elastic Net Regression"""
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=410,shuffle=True)
clf =  MOR(ElasticNet(alpha=0.01,l1_ratio=0.6,fit_intercept=True, max_iter=2000))
clf.fit(x_train,y_train)
pred = clf.predict(x_test)

pred = in_city(pred) # convert to binary predictions
y_test = in_city(y_test)

ENR_acc, ENR_f1 = accuracy_f_score(pred,y_test)
```

```python
""" Linear Regression"""
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=410,shuffle=True)
clf =  MOR(LinearRegression(n_jobs=3,normalize=False ))
clf.fit(x_train,y_train)
pred = clf.predict(x_test)

pred = in_city(pred) # convert to binary predictions
y_test = in_city(y_test)

LR_Acc, LR_f1 = accuracy_f_score(pred,y_test)
```

```python
""" Lasso Regression"""
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=410,shuffle=True)
clf =  MOR(Lasso(alpha=.5, random_state=46, fit_intercept=True))
clf.fit(x_train,y_train)
pred = clf.predict(x_test)

pred = in_city(pred) # convert to binary predictions
y_test = in_city(y_test)

lasso_Acc, lasso_f1 = accuracy_f_score(pred,y_test)
```

```python
""" Ridge Regression"""
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=410,shuffle=True)
clf =  MOR(Ridge(alpha=.5,normalize=False, solver="auto", fit_intercept=False, random_state=46 ))
clf.fit(x_train,y_train)
pred = clf.predict(x_test)

pred = in_city(pred) # convert to binary predictions
y_test = in_city(y_test)

RR_acc, RR_f1 = accuracy_f_score(pred,y_test)
```

```python
print(f"     Regression Performances   ")
print("-"*35)
print(f"Algorithm     | Accuracy | F1 Score")
print(f"XGBoost          {xg_acc}%     {xg_f1}%")
print(f"Random Forest    {rdf_acc}%     {rdf_f1}%")
print(f"Bayesian Ridge   {brr_acc}%     {brr_f1}%")
print(f"Elastic Net      {ENR_acc}%     {ENR_f1}%")
print(f"Linear           {LR_Acc}%     {LR_f1}%")
print(f"Lasso            {lasso_Acc}%     {lasso_f1}%")
print(f"Ridge            {RR_acc}%      {RR_f1}%")
```
